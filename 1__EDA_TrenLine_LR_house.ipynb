```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]

print(f"Number of outliers detected in 'price': {len(outliers)}")
print("Outliers in 'price':")
```

```python
# Filter out the outliers from the original DataFrame
df_cleaned = df[~((df['price'] < lower_bound) | (df['price'] > upper_bound))]

print(f"Original DataFrame shape: {df.shape}")
print(f"Cleaned DataFrame shape: {df_cleaned.shape}")

# Display the head of the cleaned DataFrame to verify
print("\nHead of cleaned DataFrame:")
print(df_cleaned.head())
```
