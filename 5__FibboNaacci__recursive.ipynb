```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x - 2)**2 + 1
```

```python
def fibonacci(n):

    fib = [0, 1]

    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])

    return fib
```

```python
def fibonacci_search(f, a, b, n):

    fib = fibonacci(n+1)

    k = 0

    while k < n-2:

        x1 = a + (fib[n-k-2] / fib[n-k]) * (b - a)
        x2 = a + (fib[n-k-1] / fib[n-k]) * (b - a)

        if f(x1) > f(x2):
            a = x1
        else:
            b = x2

        k += 1

    return (a + b) / 2
```

```python
xmin = fibonacci_search(f, -10, 10, 20)

print("Minimum x:", xmin)
print("Minimum value:", f(xmin))
```

```python
x = np.linspace(-10,10,200)

plt.plot(x, f(x))
plt.scatter(xmin, f(xmin), color='red')

plt.title("Fibonacci Search Optimization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
